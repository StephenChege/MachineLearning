```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
print("Modules load complete!!")
```

```python
path = '/home/vosti/machine_learning/csvs/melb_data.csv'
df = pd.read_csv(path)

df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
features = ['Landsize', 'Lattitude', 'Longtitude', 'Distance', 'Bedroom2']
X = df[features]
X.head()
```

```python
#Target
y = df['Price']
```

### **Creating Model**

```python
#Define model
model = DecisionTreeRegressor()

#fit model
model.fit(X,y)

#predict
prediction = model.predict(X.head())

print("The original prices are:", y.head().tolist())
print("The predicted prices are:", prediction)
```

The model gives an "accurate" prediction compared to the actual price but this is because we are using in-sample data for training and testing our model. Suppose we give our model a new data set that it has never seen will it give accurate predictions? 
Lets See...

```python
#Define features and target - X_test and y_test are the unseen data
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)

#fit the model
model.fit(X_train,y_train)

#predict
predict = model.predict(X_test.head())

print("The original prices are:", y.head().tolist())
print("The predicted prices are:", predict)
```

We see that the predicted prices differ from the original prices if we use new data.

So lets calculate the `mean_absolute_error` which is the absolute value of Original-predicted.

```python
#We start with the mean_absolute_error(mae) for the 'in-sample' data.
prediction_wh = model.predict(X)
mean_absolute_error(y, prediction_wh)
```

```python
#mae for unseen data
predict_wh = model.predict(X_test)
mean_absolute_error(y_test, predict_wh)
```

We see that the mae for in-sample data is lower compared to that of unseen data.

*Underfitting* is when the model fails to capture the relevant patterns and fails to give good predictions while *Overfitting* is capturing patterns that cannot be repeated again and using them for prediction thus failing to give good predictions.

Now let us try to fine tune our model by changing the tree depth using `max_leaf_nodes` argument for better perfomance.

```python
start = int(input("Enter start value:"))
stop = int(input("Enter stop value:"))
step = int(input("Enter step value:"))
max_leaf_nodes = list(range(start, stop, step))

def get_mae(max_leaf_node, X_train, X_test, y_train, y_test):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_node, random_state=0)
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, pred)
    return mae

maes = [] #list of mean absolute errors

for max_leaf_node in max_leaf_nodes:
    meaber = get_mae(max_leaf_node, X_train, X_test, y_train, y_test)
    maes.append(meaber)
    #print("Max leaf nodes %d \t MAE value: %d" %(max_leaf_node, meaber))

min_mae = min(maes) #variable for storing minimum mean absolute error
index = maes.index(min_mae) #The index of the minimum mae.
value = max_leaf_nodes[index] #The value of max_leaf_nodes with minimum Mae.
print("Max leaf nodes %d has the lowest mae of %d" %(value,min_mae))
```

After getting the number of leaf nodes with the lowest mean absolute error, we use this value to make predictions.

```python
model = DecisionTreeRegressor(max_leaf_nodes=480, random_state=0)
model.fit(X,y)
predicted_price = model.predict(X.head())
predicted_prices = model.predict(X)

pred_prices=[]
for price in predicted_price:
    vals = round(price, 1)
    pred_prices.append(vals)

print("The original prices are :", y.head().tolist())
print("The predicted prices are:", pred_prices)

#MAE
meabser = mean_absolute_error(y.head(), predicted_price)
#print("For " + str(value) + " max_leaf_nodes,", "The MAE value is: {:,.2f}".format(meabser))
print("Number of leaf nodes: %d \t MAE value: %d" %(value, meabser))
```
